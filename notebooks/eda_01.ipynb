```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# -- Load th data --
df = pd.read_csv("../data/BrentOilPrices.csv")
print(df.head)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
```

```python
df = df.sort_values(by='Date').reset_index(drop=True)
print(df.head)
```

```python
plt.figure(figsize=(15,7))
plt.plot(df['Date'], df['Price'], label='Brent oil Price (usd/barrel)')
plt.title('Historical Brent oil prices')
plt.xlabel('Date')
plt.ylabel('Price (usd/barrel)')
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df['Log_Return'] = np.log(df['Price'] / df['Price'].shift(1))
```

```python
df.dropna(subset=['Log_Return'], inplace=True)

print("\nDataFrame head with Log_Return:")
print(df.head())
print("\nDescription of Log_Return:")
print(df['Log_Return'].describe())
```

```python
# --- 5. Plots the log returns. ---
plt.figure(figsize=(15,7))
plt.plot(df['Date'], df['Log_Return'], label = 'Brent oil log returns', color='orange', alpha=0.8)
plt.title('Brent oil daily log returns')
plt.xlabel('Date')
plt.ylabel('Log_Returns')
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()

```
